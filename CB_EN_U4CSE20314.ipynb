# 4.	Perform an outlier analysis for any two columns (2) 

```python
import numpy as np
# IQR
Q1 = np.percentile(s['Distance from Residence to Work'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(s['Distance from Residence to Work'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1


# Above Upper bound
upper = s['Distance from Residence to Work'] >= (Q3+1.5*IQR)
 
print("Upper bound:",upper)
print(np.where(upper))
 
# Below Lower bound
lower = s['Distance from Residence to Work'] <= (Q1-1.5*IQR)
print("Lower bound:", lower)
print(np.where(lower))


```

```python
import numpy as np
# IQR
Q1 = np.percentile(s['Age'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(s['Age'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1


# Above Upper bound
upper = s['Age'] >= (Q3+1.5*IQR)
 
print("Upper bound:",upper)
print(np.where(upper))
 
# Below Lower bound
lower = s['Age'] <= (Q1-1.5*IQR)
print("Lower bound:", lower)
print(np.where(lower))


```
